```python
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.common.keys import Keys
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
import time

# Initialize WebDriver (replace 'chromedriver' with the path if necessary)
driver = webdriver.Chrome()

try:
    # Open the login page
    driver.get("http://127.0.0.1:8000/login-admin/")  # Update the URL to the login page

    # Maximize browser window
    driver.maximize_window()

    # Wait for the page to load
    time.sleep(2)

    # Input username/email
    username_field = driver.find_element(By.NAME, "username")  # Replace with the actual "name" attribute
    username_field.send_keys("root")  # Replace with the actual username

    # Input password
    password_field = driver.find_element(By.NAME, "password")  # Replace with the actual "name" attribute
    password_field.send_keys("1234")  # Replace with the actual password

    # Submit the login form
    password_field.send_keys(Keys.RETURN)  # Alternatively, locate the login button and click it

    # Wait for the admin dashboard to load
    WebDriverWait(driver, 10).until(
        EC.presence_of_element_located((By.LINK_TEXT, "View Details"))
    )

    # Locate "View Details" buttons
    view_details_buttons = driver.find_elements(By.LINK_TEXT, "View Details")

    # Iterate through buttons and click each one
    for button in view_details_buttons:
        button.click()

        # Wait for the next page to load
        time.sleep(2)

        # Print the current URL for confirmation
        print("Current Page URL:", driver.current_url)

        # Navigate back to the main page
        driver.back()

        # Re-locate the buttons
        time.sleep(2)
        view_details_buttons = driver.find_elements(By.LINK_TEXT, "View Details")

finally:
    # Close the browser
    driver.quit()
```

```python
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.common.keys import Keys
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
import time

# Initialize WebDriver
driver = webdriver.Chrome()

try:
    # Open the login page
    driver.get("http://127.0.0.1:8000/login-admin/")  # Replace with the actual login page URL

    # Maximize browser window
    driver.maximize_window()

    # Wait for the page to load
    time.sleep(2)

    # Input username/email (update the element locators if needed)
    username_field = driver.find_element(By.NAME, "username")  # Update with actual "name" attribute
    username_field.send_keys("root")  # Replace with actual username

    # Input password
    password_field = driver.find_element(By.NAME, "password")  # Update with actual "name" attribute
    password_field.send_keys("1234")  # Replace with actual password

    # Submit the login form
    password_field.send_keys(Keys.RETURN)

    # Wait for the admin dashboard to load
    WebDriverWait(driver, 10).until(
        EC.presence_of_element_located((By.XPATH, '/html/body/div[1]'))
    )

    print("Login successful, testing sidebar navigation.")

    # Find sidebar links
    sidebar_links = driver.find_elements(By.CSS_SELECTOR, "ul.nav_list > li > a")

    # Print sidebar link texts
    print("Sidebar Links:")
    for index, link in enumerate(sidebar_links):
        # Extract the text from the nested `span.links_name` element
        link_text = link.find_element(By.CSS_SELECTOR, "span.links_name").text.strip()
        print(f"{index}: {link_text if link_text else 'No Text Found'}")

    # Click on each link and validate the navigation
    for i in range(len(sidebar_links)):
        # Re-fetch sidebar links in each iteration
        sidebar_links = driver.find_elements(By.CSS_SELECTOR, "ul.nav_list > li > a")

        # Extract the text of the link
        link_text = sidebar_links[i].find_element(By.CSS_SELECTOR, "span.links_name").text.strip()
        print(f"Clicking link: {link_text if link_text else 'No Text Found'}")

        if link_text:  # Skip links with no text
            sidebar_links[i].click()
            time.sleep(2)

            # Print the current URL for validation
            print("Navigated to:", driver.current_url)

            # Navigate back if not the last link
            if i < len(sidebar_links) - 1:
                driver.back()
                time.sleep(2)
        else:
            print(f"Skipping link with no text at index {i}")

    # Find and click the Logout link
    logout_link = driver.find_element(By.CSS_SELECTOR, "a[href*='logout']")
    print("Clicking Logout link.")
    logout_link.click()

    # Wait for the login form to appear after logging out
    WebDriverWait(driver, 10).until(
        EC.presence_of_element_located((By.CSS_SELECTOR, "input[name='username']"))
    )
    print("Logged out successfully, test complete.")

finally:
    # Close the browser
    driver.quit()
```
